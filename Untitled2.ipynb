```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
```

```python
# Open a DataFrame and read the CSV file.
df = pd.read_csv('/content/framingham.csv')

# show the DataFrame
df
```

```python
import matplotlib.pyplot as plt

def plot_mean_cigarettes_per_day_by_age(data):
    """
    Plot the mean cigarettes per day by age for individuals aged 32 to 50.

    Parameters:
    - data (DataFrame): Input DataFrame containing the data.

    Returns:
    - None
    """
    # Filter data for individuals aged 32 to 50
    data_filtered = data[(data['age'] >= 32) & (data['age'] <= 50)]

    # Remove NaN values from the 'cigsPerDay' column.
    data_filtered = data_filtered.dropna(subset=['cigsPerDay'])

    # Calculate the average daily cigarette intake by grouping data by age.
    age_cigs_mean = data_filtered.groupby('age')['cigsPerDay'].mean().reset_index()

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.bar(age_cigs_mean['age'], age_cigs_mean['cigsPerDay'], color='lightgreen')
    plt.xlabel('Age', fontsize=14)
    plt.ylabel('Cigarettes Per Day (Mean)', fontsize=14)
    plt.title('Mean Cigarettes Per Day by Age (Aged 32 to 50)')
    plt.xticks(range(32, 51), rotation=360, fontsize=14)
    plt.yticks(fontsize=14)

    # Rotate x-labels for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Assuming 'df' is your DataFrame
plot_mean_cigarettes_per_day_by_age(df)
```

```python
def plot_age_vs_heart_rate(data):
    """
    Plot a scatter plot of age vs. heart rate using the first 2000 rows of the input DataFrame.

    Parameters:
    - data (DataFrame): Input DataFrame containing the data.

    Returns:
    - None
    """
    # Take the first 2000 rows and 16 columns
    subset_data = data.iloc[:2000, :16]

    # Plotting the data
    plt.figure(figsize=(10, 6))

    # Indicate which columns you wish to plot.
    x = subset_data['age']
    y = subset_data['heartRate']

    # Create the scatter plot
    plt.scatter(x, y, color='blue', alpha=0.5)

    # Add title and labels
    plt.xlabel('Age', fontsize=14)
    plt.ylabel('Heart Rate', fontsize=14)
    plt.title('Scatter Plot: Age vs. Heart Rate (First 2000 rows)', fontsize=14)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)

    # Show plot
    plt.grid(True)
    plt.show()

# Assuming 'df' is your DataFrame
plot_age_vs_heart_rate(df)
```

```python
def plot_male_smoking_distribution(data):
    """
    Plot the distribution of smokers and non-smokers among males using a pie chart.

    Parameters:
    - data (DataFrame): Input DataFrame containing the data.

    Returns:
    - None
    """
    # Select data only from male
    male_data = data[data['male'] == 1]

    # Determine how many male smoke (currentSmoker = 1) and how many male do not smoke (currentSmoker = 0).
    male_smokers = male_data[male_data['currentSmoker'] == 1].shape[0]
    male_non_smokers = male_data[male_data['currentSmoker'] == 0].shape[0]

    # Create a pie chart
    labels = ['Smokers', 'Non-smokers']
    sizes = [male_smokers, male_non_smokers]
    colors = ['lightcoral', 'lightskyblue']
    explode = (0.1, 0)  # explode the smokers slice to highlight it

    plt.figure(figsize=(8, 6))
    plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
    plt.title('Distribution of Smokers and Non-smokers Among Males')
    # Draw a pie as a circle is guaranteed by an equal aspect ratio.
    plt.axis('equal')
    plt.show()

# Assuming 'df' is your DataFrame
plot_male_smoking_distribution(df)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Preprocess the data to handle NaN or infinite values
df.dropna(subset=['BMI', 'totChol'], inplace=True)
inf_values = df[['BMI', 'totChol']].isin([np.inf, -np.inf]).any(axis=1)
data = df[~inf_values]

# Selecting the features for clustering
X = data[['BMI', 'totChol']]

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Finding the optimal number of clusters using the Elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow method to visualize the optimal number of clusters
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-')
plt.title('Elbow Method for BMI and Cholesterol Clustering')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')  # Within-cluster sum of squares
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()

# Based on the Elbow method, choose the optimal number of clusters and apply K-means clustering
optimal_num_clusters = 3  # You can choose the optimal number of clusters based on the Elbow plot
kmeans = KMeans(n_clusters=optimal_num_clusters, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Add the cluster labels to the dataset
data['Cluster'] = clusters

# Visualizing the clusters with predicted points
plt.figure(figsize=(10, 6))
plt.scatter(data['BMI'], data['totChol'], c=data['Cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='*', label='Centroids')
plt.title('K-means Clustering with Predicted Points')
plt.xlabel('BMI')
plt.ylabel('Total Cholesterol')
plt.legend()
plt.grid(True)

# Example points to predict their clusters
np.random.seed(0)
num_points = 3
random_bmi = np.random.uniform(data['BMI'].min(), data['BMI'].max(), num_points)
random_chol = np.random.uniform(data['totChol'].min(), data['totChol'].max(), num_points)
points_to_predict = np.column_stack((random_bmi, random_chol))

# Standardize the points using the same scaler
points_scaled = scaler.transform(points_to_predict)

# Predict the cluster of the points
predicted_clusters = kmeans.predict(points_scaled)

# Plotting the random points with predicted clusters
plt.scatter(points_to_predict[:, 0], points_to_predict[:, 1], c=predicted_clusters, cmap='cool', marker='x', label='Random Points')
for i, txt in enumerate(predicted_clusters):
    plt.annotate(txt, (points_to_predict[i, 0], points_to_predict[i, 1]), textcoords="offset points", xytext=(0,10), ha='center')
plt.show()
```

```python
# Drop rows with NaN values in 'BMI' and 'totChol'
data.dropna(subset=['BMI', 'totChol'], inplace=True)

# Selecting the features for line fitting
X = data['BMI'].values.reshape(-1, 1)  # Reshape to a 2D array
y = data['totChol']

# Impute missing values with mean (if any)
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Plotting the line fitting
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Original data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Line fitting')
plt.title('Line Fitting for BMI and Total Cholesterol')
plt.xlabel('BMI')
plt.ylabel('Total Cholesterol')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Select the desired columns
selected_columns = ['male', 'age', 'currentSmoker', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose', 'TenYearCHD']

# Create DataFrame df_new with selected columns
df_new = df[selected_columns]
def plot_heatmap(df_new):
    """
    Create a heatmap to observe the correlation between attributes.

    Parameters:
    - data (DataFrame): Input DataFrame containing the data.

    Returns:
    - None
    """
    plt.figure(figsize=(10, 8), dpi=80)
    # Calculate correlation matrix for numeric columns only
    corr_matrix = df_new.select_dtypes(include='number').corr()
    # Plotting a heatmap
    sns.heatmap(corr_matrix, annot=True, cmap='Blues', linewidths=.5)
    plt.title('Correlation between various factors')
    # Rotate x-axis labels if needed
    plt.xticks(rotation=90, ha='right', fontsize=14)
    # Rotate y-axis labels if needed
    plt.yticks(rotation=360, fontsize=14)
    # Save and display the plot
    plt.savefig('Heatmap.png')
    plt.show()

# Assuming 'data' is your DataFrame
plot_heatmap(df_new)
```
